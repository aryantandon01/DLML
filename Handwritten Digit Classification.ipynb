```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
x_train_flattened = x_train.reshape(len(x_train), 28*28)
# x_train.reshape(60000, 28*28) can also be written
x_train_flattened
```

```python
x_train_flattened.shape
```

```python
x_test_flattened = x_test.reshape(len(x_test), 28*28)
# x_test.reshape(60000, 28*28) can also be written
x_test_flattened
```

```python
# To scale the pixel values from 0 to 1
x_train_flattened = x_train_flattened / 255
x_test_flattened = x_test_flattened / 255
```

```python
x_test_flattened.shape
```

```python
x_train_flattened[0]
```

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape = (784,), activation = 'sigmoid')
    # keras.layers.Dense(output layers, input layers, activation func) 
])

model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)

model.fit(x_train_flattened, y_train, epochs = 5)
```

```python
model.evaluate(x_test_flattened, y_test)
```

```python
plt.matshow(x_test[5])
```

```python
y_predicted = model.predict(x_test_flattened)
y_predicted[5]
# Output shows the probab of each digit being the one in the picture
```

```python
np.argmax(y_predicted[5])
```

```python
y_predicted_labels = [np.argmax(i) for i in y_predicted]
```

```python
y_predicted_labels
```

```python
cm = tf.math.confusion_matrix(labels = y_test, predictions = y_predicted_labels)
cm
```

```python
plt.figure(figsize = (10, 8))
sns.heatmap(cm, annot = True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
model = keras.Sequential([
    keras.layers.Dense(100, input_shape = (784,), activation = 'relu'),   # Adding more output layers
    keras.layers.Dense(10, activation = 'sigmoid')
])

model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)

model.fit(x_train_flattened, y_train, epochs = 5)
```
